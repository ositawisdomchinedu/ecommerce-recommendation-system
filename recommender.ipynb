### Import Labraries

```python
from sentence_transformers import SentenceTransformer
import numpy as np
import pandas as pd
import os
import glob
```

### Load Model

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
```

### Load Data

```python
def load_all_csv(folder_path):
    all_files = glob.glob(os.path.join(folder_path, "*.csv"))
    
    df_list = []
    
    for file in all_files:
        try:
            temp_df = pd.read_csv(file)
            temp_df['source_file'] = os.path.basename(file)
            df_list.append(temp_df)
        except Exception as e:
            print(f"Error loading {file}: {e}")
    
    df = pd.concat(df_list, ignore_index=True, sort=False)
    return df

# Load data
df = load_all_csv("data")
```

```python
# Fill numeric columns with median
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# Fill text columns with empty string
cat_cols = df.select_dtypes(include=['object']).columns
df[cat_cols] = df[cat_cols].fillna('')
```

```python
df.isnull().sum()
```

```python
# Drop duplicates
df = df.drop_duplicates(subset='product_id')
```

### Feature Engineering

```python
# create combined features for embedding
df['combined_features'] = (
    df['title'].astype(str) + " " +
    df['category'].astype(str) + " " +
    df['product_description'].astype(str)
)
```

### Generate Embeddings

```python
embeddings = model.encode(
    df['combined_features'].tolist(),
    show_progress_bar=True
)
```

### Map index with Product_id

```python
id_to_index = pd.Series(df.index, index=df['product_id'])
```

### Create Recommendation Function

```python
from sklearn.metrics.pairwise import cosine_similarity

def recommend_products(product_id, top_n=5):
    if product_id not in id_to_index:
        return "Product not found"
    
    idx = id_to_index[product_id]
    
    query_vec = embeddings[idx].reshape(1, -1)
    
    scores = cosine_similarity(query_vec, embeddings).flatten()
    
    top_indices = np.argsort(scores)[::-1][1:top_n+1]
    
    return df.iloc[top_indices][
        ['product_id', 'title', 'product_description', 'category']
    ]
```

### Create Semantic Search Function

```python
def semantic_search(query, top_n=5, threshold=0.3):
    query_vec = model.encode([query])
    
    scores = cosine_similarity(query_vec, embeddings).flatten()
    
    top_indices = np.argsort(scores)[::-1][:top_n]
    
    results = df.iloc[top_indices].copy()
    results['score'] = scores[top_indices]
    
    # Filter weak matches
    results = results[results['score'] > threshold]
    
    return results.sort_values(by='score', ascending=False)
```

```python
def recommend_from_query(query, top_n=5):
    results = semantic_search(query, top_n=3)
    
    if results.empty:
        return f"No relevant products found for '{query}'"
    
    selected = results.iloc[0]
    
    print(f"Selected Product: {selected['title']}")
    print(f"Category: {selected['category']}")
    
    return recommend_products(selected['product_id'], top_n)
```
